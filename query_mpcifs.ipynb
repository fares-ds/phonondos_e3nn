```python
from pymatgen.ext.matproj import MPRester
from pymatgen.core.structure import Structure

import pandas as pd
USER_API_KEY = 'YOUR_API_KEY'
```

```python
icsd_pd = pd.read_csv('data/mp_data.csv')
```

```python
for i in range(len(icsd_pd)):
    material_id = icsd_pd.iloc[i,0]
    with MPRester(USER_API_KEY) as m:
        cif_data = m.query(criteria={'material_id': material_id}, properties=['cif'])
    if len(cif_data) > 0:
        print('Downloading task: {:5d}    '.format(i) + 'Material ID: '+material_id)
        with open('data/'+material_id+'.cif', 'w') as f:
            f.write(cif_data[0]['cif'])
    else:
        print('Skipping task: {:5d}    '.format(i) + 'Material ID: '+material_id)
```
